```python
# Analyze trends using groupby() + NumPy on a multi-level categorical dataset
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'Region': ['North', 'North', 'South', 'South', 'North', 'South'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 80, 90, 120, 85],
    'Profit': [20, 30, 15, 18, 22, 17]
})

grouped = df.groupby(['Region', 'Product'])

summary = grouped.agg(np.mean)

print(summary)
```

```python
# Create a summary DataFrame with a correlation matrix using Pandas and NumPy
```

```python
correlation_matrix = df[['Sales', 'Profit']].corr()

data = df[['Sales', 'Profit']].to_numpy()
np_corr_matrix = np.corrcoef(data.T)

np_corr_df = pd.DataFrame(np_corr_matrix, columns=['Sales', 'Profit'], index=['Sales', 'Profit'])

print("Pandas Correlation Matrix:\n", correlation_matrix)
print("\nNumPy Correlation Matrix:\n", np_corr_df)
```

```python
# Perform rolling mean on a time series and visualize using Matplotlib
```

```python
dates = pd.date_range(start='2024-01-01', periods=10, freq='D')
sales = [100, 110, 105, 115, 120, 130, 125, 140, 135, 150]

df_time = pd.DataFrame({'Date': dates, 'Sales': sales})
df_time.set_index('Date', inplace=True)

df_time['Rolling_Mean'] = df_time['Sales'].rolling(window=3).mean()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_time.index, df_time['Sales'], label='Original Sales', marker='o')
plt.plot(df_time.index, df_time['Rolling_Mean'], label='Rolling Mean (3)', linestyle='--')
plt.title('Sales Trend with Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
